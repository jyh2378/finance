```python
def get_upside_potential(value_summary):
    """
    Calculate the upside potential of a stock based on its fair price.
    """
    period = "(This Year)"  # "(This Year)", "(Next Year)"

    upside_potential = (value_summary[f"Forward PEGR Fair Price {period}"] / value_summary["Price"]) - 1
    value_summary["Upside Potential"] = upside_potential
    return value_summary
```
